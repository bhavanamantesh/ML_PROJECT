```python
#step 1 create a small dataset
import pandas as pd
data={
    "size_sqft":[500,800,1000,1200,1500,1800,2000],
    "price":[15,25,30,36,45,54,60]
}
df=pd.DataFrame(data)
df

```

```python
#step2 features and target
X=df[["size_sqft"]]
y=df["price"]
```

```python
#step3 train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(
    X,y,test_size=0.20,random_state=42
)
```

```python
#train the model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
new_house=[[2200]]
predicted_price=model.predict(new_house)
print("pridected price:",predicted_price[0],"lakhs")
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
plt.plot(X,model.predict(X))
plt.xlabel("size sqft")
plt.ylabel("price(lakhs)")
plt.title("graph for house price prediction")
plt.show()
```

```python
import pandas as pd
df=pd.read_csv("data_set1 (1).csv")
df.head()
```

```python
df.shape
```

```python
duplicated=df.duplicated()
df.duplicated().sum()
```

```python
#removeduplicate
df.drop_duplicates(inplace=True)
```

```python
#to check is null
df.isnull().sum()
```

```python
#handling null values
#numerical columns.....>median
#with for loop
num_cols=["Age","Marks","Attendance_Percent"]
for col in num_cols:
  df[col]=df[col].fillna(df[col].median())
```

```python
#handling null values
#numerical columns.....>median
#without for loop
num_cols=["Age","Marks","Attendance_Percent"]
df[num_cols]=df[num_cols].fillna(df[num_cols].median())
```

```python
#handling null values
#categorical columns.....>mode
#with for loop
cat_cols=["Gender","Department","City"]
for col in cat_cols:
  df[col]=df[col].fillna(df[col].mode()[0])
```

```python
#handling null values
#categorical columns.....>mode
#without for loop
cat_cols=["Gender","Department","City"]
df[cat_cols]=df[cat_cols].fillna(df[cat_cols].mode())
```

```python
df['Gender']=df['Gender'].replace({
    "M":"Male",
    "male":"Male",
    "m":"Male",
    "F":"Female",
    "female":"Female",
    "f":"Female",
})
print(df)
```

```python
df['Fee_Paid']=df['Fee_Paid'].replace({
    "n":"NO",
    "N":"NO",
    "Y":"YES",
    "y":"YES",
})
print(df)
```

```python
df['Email']=df['Email'].fillna("unknown")
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Course_Level']=le.fit_transform(df['Course_Level'])
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Department']=le.fit_transform(df['Department'])
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
method=['Course_Level','Department']
for cols in method:
  df[cols]=le.fit_transform(df[cols])
df.head()
```

```python
#binary encoding
df['Scholarship'].unique()
df['Scholarship']=df['Scholarship'].map({"Yes":1,"No":0})
df.head()
```
