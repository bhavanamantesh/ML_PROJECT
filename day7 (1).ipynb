```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
data={
    'sr_no':['C1','C2','C3','C4','C5','C6','C7'],
     'age':[20,40,30,18,28,35,45],
     'amount':[500,1000,800,300,1200,1400,1800]
}
df=pd.DataFrame(data)
print(df)
```

```python
#select features
X=df[['age','amount']]
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
#apply Kmeans
kmeans=KMeans(n_clusters=2,random_state=42)
df['Cluster']=kmeans.fit_predict(X_scaled)
print(df)
```

```python
import numpy as np
initial_centroids=np.array([
    [30,800],#c3
    [18,300]#c4
    ])
kmeans=KMeans(n_clusters=2,init=initial_centroids,n_init=1)
df['Cluster']=kmeans.fit_predict(df[['age','amount']])
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
data={
    'sr_no':['C1','C2','C3','C4','C5','C6','C7','C8'],
     'age':[22,25,47,52,23,40,36,28],
     'anual_income':[25000,32000,58000,60000,27000,52000,49000,31000],
     'score':[40,60,30,35,75,50,55,80]
}
df=pd.DataFrame(data)
print(df)

#select features
X=df[['age','anual_income','score']]
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
#apply Kmeans
kmeans=KMeans(n_clusters=2,random_state=42)
df['Cluster']=kmeans.fit_predict(X_scaled)
print(df)

import numpy as np
initial_centroids=np.array([
    [47,58000,30],#c3
    [52,60000,35]#c4
    ])
kmeans=KMeans(n_clusters=2,init=initial_centroids,n_init=1)
df['Cluster']=kmeans.fit_predict(df[['age','anual_income','score']])
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
data={
    'sr_no':['C1','C2','C3','C4','C5','C6','C7','C8'],
     'study_hrs':[2,5,8,1,6,3,7,4],
     'attendence':[60,75,90,55,80,65,88,70],
     'marks':[45,65,85,40,70,50,82,60]
}
df=pd.DataFrame(data)
print(df)

#select features
X=df[['study_hrs','attendence','marks']]
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
#apply Kmeans
kmeans=KMeans(n_clusters=2,random_state=42)
df['Cluster']=kmeans.fit_predict(X_scaled)
print(df)

import numpy as np
initial_centroids=np.array([
    [8,90,85],#c3
    [1,55,40]#c4
    ])
kmeans=KMeans(n_clusters=2,init=initial_centroids,n_init=1)
df['Cluster']=kmeans.fit_predict(df[['study_hrs','attendence','marks']])
print(df)
```
